```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'firstname': ['ROHIT','SAHIL','RAHUL','KOMAL','VIKRAM','POOJA','DEEPAK','KAVYA','MANISH','RITU','SURESH','DIVYA'],
    'lastname':  ['YADAV','YADAV','SHARMA','YADAV','GUPTA','YADAV','VERMA','GUPTA','SHARMA','VERMA','GUPTA','SHARMA'],
    'math':    [85,72,91,68,88,76,55,93,61,79,84,70],
    'science': [78,85,74,90,65,88,72,81,69,95,77,83],
    'english': [90,68,82,75,71,94,80,76,88,65,73,91]
}
df = pd.DataFrame(data)
df['total'] = df['math'] + df['science'] + df['english']
print(df)
```

```python
lastname_total = df.groupby('lastname')['total'].sum()
plt.figure(figsize=(8,8))
plt.pie(lastname_total, labels=lastname_total.index, autopct='%1.1f%%', wedgeprops=dict(width=0.55))
plt.title('Total Marks by Last Name')
plt.show()
```

```python
df_sorted = df.sort_values('total', ascending=True)
plt.figure(figsize=(10,7))
plt.barh(df_sorted['firstname'], df_sorted['total'])
plt.title('Total Marks per Student')
plt.xlabel('Total Marks')
plt.show()
```

```python
plt.figure(figsize=(13,6))
plt.plot(df['firstname'], df['math'],    marker='o', label='Math')
plt.plot(df['firstname'], df['science'], marker='s', label='Science')
plt.plot(df['firstname'], df['english'], marker='^', label='English')
plt.xticks(rotation=45)
plt.title('Subject-wise Marks per Student')
plt.legend()
plt.grid(True)
plt.show()
```

```python
heat_data = df.set_index('firstname')[['math','science','english']]
plt.figure(figsize=(8,9))
sns.heatmap(heat_data, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Student x Subject Heatmap')
plt.show()
```
